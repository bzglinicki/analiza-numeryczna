1. Przeanalizuj funkcję `int_lagrange`, przyjmującą jako argumenty dwie listy liczb, $X$ oraz $Y$ oraz liczbę rzeczywistą $x$. Funkcja wyznacza wartość wielomianu interpolacyjnego Lagrange'a w punkcie $x$, otrzymanego na podstawie punktów $(X, Y)$, posługując się bazą Newtona i wykorzystując algorytm różnic dzielonych.

```python
def divided_diff(X: list[float], Y: list[float]) -> np.ndarray:
    '''
    Oblicza tablicę różnic dzielonych.
    '''
    n = len(Y)
    coef = np.zeros([n, n])
    
    # Pierwsza kolumna to po prostu wartości funkcji w punktach X.
    coef[:,0] = Y
    
    for j in range(1,n):
        for i in range(n-j):
            coef[i][j] = (coef[i+1][j-1] - coef[i][j-1]) / (X[i+j]-X[i])
            
    return coef
```

```python
def int_lagrange(X: list[float], Y: list[float], x: float) -> float:
    """Oblicza wartość wielomianu interpolacyjnego Lagrange'a w punkcie x na podstawie danych (X, Y)."""

    coef = divided_diff(X, Y)[0, :]
    n = len(X) - 1 

    y = coef[n]
    for k in range(1, n+1):
        y = coef[n-k] + (x -X[n-k])*y
    
    return y
```
