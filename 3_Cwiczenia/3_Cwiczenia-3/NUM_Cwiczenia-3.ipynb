### Zadanie 1. Sortowanie.

*Sortowanie* to zadanie polegające na uporządkowaniu elementów pewnego zbioru, na którym określona jest relacja porządku. Sortować możemy więc na przykład liczby rzeczywiste, posługując się naturalnym porządkiem $\leq $, oraz słowa, wykorzystując porządek alfabetyczny.

Istnieje wiele algorytmów sortowania. Przyjrzymy się trzem z nich:
- sortowanie przez wybieranie (ang. *selection sort*),
- sortowanie szybkie (ang. *quick sort*),
- sortowanie przez scalananie (ang. *merge sort*).

2. Poniżej znajdziesz przykładowe implementacje każdego z tych algorytmów. Funkcje te przyjmują jako argument jednowymiarową tablicę NumPy i zwracają posortowaną tablicę o tych samych elementach. Przyjrzyj się im i postaraj się je zrozumieć.

````python
def selection_sort(arr: np.ndarray) -> None:
    """
    Sortuje tablicę NumPy in-place za pomocą algorytmu selection sort.
    """
    n = len(arr)
    for i in range(n):
        # Zakładamy, że minimalny element jest na pozycji i
        min_idx = i
        
        # Szukamy najmniejszego elementu w pozostałej części tablicy
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        
        # Zamieniamy elementy miejscami
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
````

````python
# Przykład użycia:
a = np.array([64, 25, 12, 22, 11])
selection_sort(a)
print(a)  # wynik: [11 12 22 25 64]
````

````python
def merge_sort(arr: np.ndarray) -> None:
    """
    Sortuje tablicę NumPy in-place za pomocą algorytmu merge sort.
    """
    def _merge_sort(a, left, right):
        if left < right:
            mid = (left + right) // 2
            _merge_sort(a, left, mid)
            _merge_sort(a, mid + 1, right)
            _merge(a, left, mid, right)

    def _merge(a, left, mid, right):
        # Tworzymy kopie podtablic
        left_part = a[left:mid+1].copy()
        right_part = a[mid+1:right+1].copy()

        i = j = 0
        k = left

        # Scalanie dwóch posortowanych części
        while i < len(left_part) and j < len(right_part):
            if left_part[i] <= right_part[j]:
                a[k] = left_part[i]
                i += 1
            else:
                a[k] = right_part[j]
                j += 1
            k += 1

        # Jeśli zostały elementy w lewej części
        while i < len(left_part):
            a[k] = left_part[i]
            i += 1
            k += 1

        # Jeśli zostały elementy w prawej części
        while j < len(right_part):
            a[k] = right_part[j]
            j += 1
            k += 1

    _merge_sort(arr, 0, len(arr) - 1)
````

````python
# Przykład użycia:
a = np.array([38, 27, 43, 3, 9, 82, 10])
merge_sort(a)
print(a)  # wynik: [ 3  9 10 27 38 43 82 ]
````

````python
def quick_sort(arr: np.ndarray) -> None:
    """
    Sortuje tablicę NumPy in-place za pomocą algorytmu quick sort.
    """
    def _quick_sort(a, low, high):
        if low < high:
            pivot_index = _partition(a, low, high)
            _quick_sort(a, low, pivot_index - 1)
            _quick_sort(a, pivot_index + 1, high)

    def _partition(a, low, high):
        pivot = a[high]  # wybieramy ostatni element jako pivot
        i = low - 1

        for j in range(low, high):
            if a[j] <= pivot:
                i += 1
                a[i], a[j] = a[j], a[i]

        # umieszczamy pivot na właściwym miejscu
        a[i + 1], a[high] = a[high], a[i + 1]
        return i + 1

    _quick_sort(arr, 0, len(arr) - 1)
````

````python
# Przykład użycia:
a = np.array([10, 7, 8, 9, 1, 5])
quick_sort(a)
print(a)  # wynik: [1 5 7 8 9 10]
````

### Zadanie 3. Statystyka tekstu.

Poniższy kod tworzy plik `nowela.txt` zawierający losową liczbę automatycznie wygenerowanych zdań w języku polskim (niekoniecznie sensownych).

````python
import random

# Elementy zdań
subjects = [
    "Kot", "Pies", "Programista", "Nauczyciel", "Student",
    "Moja siostra", "Mój sąsiad", "Samochód", "Komputer", "Dziecko"
]

verbs = [
    "lubi", "czyta", "pisze", "ogląda", "tworzy",
    "kupuje", "sprząta", "odkrywa", "buduje", "analizuje"
]

objects = [
    "książki", "programy komputerowe", "ciekawe filmy", "nowe technologie",
    "smaczne obiady", "stare fotografie", "trudne zadania",
    "piękne krajobrazy", "różne projekty", "muzykę klasyczną"
]

# Losowanie liczby zdań do wygenerowania - między 3 a 10
num_sentences = random.randint(3, 10)

# Generowanie zdań
sentences = []
for _ in range(num_sentences):
    subject = random.choice(subjects)
    verb = random.choice(verbs)
    obj = random.choice(objects)

    sentence = f"{subject} {verb} {obj}."
    sentences.append(sentence)

# Zapis do pliku
with open("nowela.txt", "w", encoding="utf-8") as file:
    for sentence in sentences:
        file.write(sentence + " ")

print("Plik nowela.txt został utworzony.")
````

**CSV** (ang. *Comma-separated values* - *wartości oddzielone przecinkami*) to tekstowy format przechowywania danych tabelarycznych. Kolejne wiersze pliku w formacie CSV odpowiadają kolejnym wierszom tabeli (jednak pierwszy wiersz pliku CSV zawiera zwykłe nagłówki kolumn), zaś w każdym wierszu dane z kolejnych kolumn oddzielone są przecinkami (lub innymi, z góry określonymi separatorami, np. średnikami lub tabulatorami).

